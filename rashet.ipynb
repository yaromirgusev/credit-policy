Оценка влияния интруменов кредитной политики на экономический рост России

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('dataset.xlsx')
data = data.drop('Год', axis=1)
data
```

```python
data.info()
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
X = sm.add_constant(np.column_stack((data.X1, data.X2, data.X3, data.X4, data.X5)))
Y = np.array(data.Y)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
X = sm.add_constant(np.column_stack((data.X1, data.X2, data.X3, data.X4)))
Y = np.array(data.Y)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
X = sm.add_constant(np.column_stack((data.X1, data.X2, data.X3)))
Y = np.array(data.Y)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
X = sm.add_constant(np.column_stack((data.X1, data.X2)))
Y = np.array(data.Y)
model = sm.OLS(Y,X).fit()
model.summary()
```

```python
from statsmodels.stats.diagnostic import het_white, acorr_breusch_godfrey
from statsmodels.stats.stattools import jarque_bera
residuals = model.resid
white_test = het_white(residuals,X)
print("Тест Уайта")
print("LM Statistic:", white_test[0])
print("LM-Test p-value:", white_test[1])
print("F-Statistic:", white_test[2])
print("F-Test p-value:", white_test[3],'\n')
```

```python
standardized_residuals = model.get_influence().resid_studentized_internal

sm.qqplot(standardized_residuals, line='45', fit=True)
plt.title('Q-Q Plot остатков')
plt.show()
```

```python
plt.hist(residuals, bins = 20);
```

```python
import statsmodels.api as sm

dw_test_statistic = sm.stats.durbin_watson(model.resid)
dw_test_statistic
```
